```python
import matplotlib.pyplot as plt
import numpy as np
```

If we specify only one set of values in plot() function as a parameter, Matplotlib will by-default understand as Y-Axis values and will plot the graph with X-Axis values beign the same but starting with 0. Example:
plot([1,2,3,4]) --> This will be the value of Y axis as by Matplotlib
So the X-Axis values will be --> ([0,1,2,3])

Below is the Coding

```python
plt.plot([1,2,3,4,5])
plt.ylabel('Some Values')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.axis([0,6,0,20])
```

Above mentioned axis() method is used to set custom axis range values for both Axes.
It takes 4 Arguments and it is in the form of [xmin,xmax,ymin,ymax] in the above example --
[0,6,0,20] 0 -> min value for X axis. 6 -> max value for X axis. 
            0 -> min value for Y axis. 20 -> max value for Y axis.

```python
a=[1,9,3,8]
b=[4,5,6,7]
plt.plot(a, b)
```

There are various plotting techniques we can use to make our plot more intuitive, formatting the style of pur plots.
There are various methods by which we can implement various types of Plotting, some of them are shown Below:-

```python
plt.plot(a,b,'ro')
```

Above is a plot with method 'ro' which indicates as --> 'r' for red and 'o' for circle

```python
plt.plot(a,b,'go--',linewidth=2,markersize=6)
```

Above is an example of 'go--' methods where 'g' is green, 'o' is circle,'--' is dashed line
markersize is the size of the circle and linewidth is the with of the dashed line

Matplotlib is not just limited to plot two variables values given in list form, we can also use arrays to plot on the graph.

```python
t=np.arange(0,5,0.2)
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
```

```python
data={'a':np.arange(50),
'c':np.random.randint(0,50,50),
'd':np.random.randn(50)}
data['b']=data['a']+10*np.random.randn(50)
data['d']=np.abs(data['d'])*100
plt.scatter('a','b',c='c',s='d',data=data) 
# c is the color acc to the variable defined and s is the size acc to the variable defined
# Default color 'c' is RGB with varying shades. 'alpha' is a keyword for the hue of the color
plt.xlabel('Entry A')
plt.ylabel('Entry B')
plt.show()
```

```python
# Plotting with Categorical Variables
names=['group_a','group_b','group_c']
values=[1,20,100]
plt.figure(figsize=(9,4),layout='constrained')
plt.subplot(131)
plt.bar(names,values)
plt.subplot(132)
plt.bar(names,values)
plt.subplot(133)
plt.bar(names,values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```
