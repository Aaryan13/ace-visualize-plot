```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Plotting two basic variables
x=[5, 2, 9, 4, 7]
y=[10, 5, 8, 4, 2]
plt.plot(x,y)
```

```python
# Bar Graph Plot
plt.bar(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.axis((0,10,0,20))
```

```python
a1=[10,20,30,40]
b1=[20,25,35,55]
plt.plot(a1,b1)
plt.title("Linear Graph",fontsize=18,color="#FF6E31")
```

```python
a2=[4,5,8,9,11]
b2=[1,6,9,7,10]
plt.plot(a2,b2)
plt.title("Linear Graph",fontsize=18,color="orange")
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")
plt.ylim(0,14)
xlabel=np.arange(0,11)
plt.xticks(xlabel)
plt.show()
```

```python
a3=[10, 20, 30, 40]
b3=[20, 25, 35, 55]
fig=plt.figure(figsize=(7,4), facecolor='green',edgecolor='blue',linewidth=8)
ax=fig.add_axes((1,1,0.7,1)) #axes([left,bottom,height,width]) all between fractions of 0 & 1
ax.plot(a3,b3)
plt.title("Linear Graph",color='blue',fontsize=18)
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.legend(["Line"],loc='upper left',facecolor='pink')
```

```python
a4=[10,20,30,40]
b4=[20,25,35,55]
fig=plt.figure(figsize=(6,4))
plt.plot(a4)
plt.plot(b4)
plt.title("Two Line Plot")
plt.legend(["Line 1","Line 2"])
```

```python
a5=[10, 20, 30, 40]
b5=[20, 25, 35, 55]
plt.figure()
plt.subplot(121)
plt.plot(a5,b5)
plt.subplot(122)
plt.plot(b5,a5)
```

```python
a6=[2,5,8,9,7]
b6=[1,2,7,8,9]
fig,axes=plt.subplots(1,2)
axes[0].plot(a6,b6)
axes[1].plot(b6,a6)
```

```python
a7=[12,4,8,3,6]
b7=[10,11,9,7,6]
plt.plot(b7, a7,'o--')
plt.title("Line Chart")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

```python
a8=[5,6,8,4,3]
b8=[9,6,2,4,8]
plt.plot(a8,b8,color='#13005A',linewidth=2,marker='o',markersize=8,linestyle='--',markerfacecolor='#8DCBE6',markeredgecolor='red')
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")
plt.title("Line Chart")
plt.xticks(np.arange(0,10))
plt.yticks(np.arange(0,10))
plt.show()
```

```python
year=[1972,1982,1992,2002,2012]
e_india=[100.6,158.61,305.54,394.96,724.79]
de_nepal=[10.5,25.21,58.65,119.27,274.87]
plt.plot(year, e_india,color='#301E67',marker='o',markersize=7,markerfacecolor='#F48484',label='India')
plt.plot(year, de_nepal,color='#13005A',marker='o',markerfacecolor='#F273E6',markersize='7',linestyle='dashed',label='Nepal')
plt.xlabel('Years')
plt.ylabel('Power Consumption in kWh')
plt.title('Electricity Consumption per Capita of India and Nepal')
plt.legend()
```

```python
n = 50
x = np.random.rand(n) * 10
y = np.random.rand(n) * 10
colors = ['red', 'blue', 'green', 'purple', 'orange']
point_colors = np.random.choice(colors, size=n)
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=point_colors, s=100)
plt.title('Simple Colorful Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
