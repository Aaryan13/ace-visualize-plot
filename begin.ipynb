```python
year=[1972,1982,1992,2002,2012]
e_india=[100.6,158.61,305.54,394.96,724.79]
e_bangla=[10.5,25.21,58.65,119.27,274.87]
plt.plot(year, e_india,color='#301E67',label='India')
plt.plot(year, e_bangla,color='#8EA7E9',label='Bangladesh')
plt.xlabel('Years')
plt.ylabel('Power Consumption in kWh')
plt.title('Electricity Consumption per Capita of India and Bangladesh')
plt.legend()
```

```python
year=[1972,1982,1992,2002,2012]
e_india=[100.6,158.61,305.54,394.96,724.79]
e_bangla=[10.5,25.21,58.65,119.27,274.87]
plt.plot(year, e_india,color='#301E67',marker='o',markersize=7,markerfacecolor='#F48484',label='India')
plt.plot(year, e_bangla,color='#13005A',marker='o',markerfacecolor='#F273E6',markersize='7',linestyle='dashed',label='Bangladesh')
plt.xlabel('Years')
plt.ylabel('Power Consumption in kWh')
plt.title('Electricity Consumption per Capita of India and Bangladesh')
plt.legend()
```

```python
a1=[10,20,30,40]
b1=[20,25,35,55]
plt.plot(a1,b1)
plt.title("Linear Graph",fontsize=18,color="#FF6E31")
```

```python
a2=[4,5,8,9,11]
b2=[1,6,9,7,10]
plt.plot(a2,b2)
plt.title("Linear Graph",fontsize=18,color="orange")
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")
plt.ylim(0,14)
xlabel=np.arange(0,11)
plt.xticks(xlabel)
plt.show()
```

```python
a3=[10, 20, 30, 40]
b3=[20, 25, 35, 55]
fig=plt.figure(figsize=(7,4), facecolor='green',edgecolor='blue',linewidth=8)
ax=fig.add_axes([1,1,0.7,1]) #axes([left,bottom,height,width]) all between fractions of 0 & 1
ax.plot(a3,b3)
plt.title("Linear Graph",color='blue',fontsize=18)
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.legend(["Line"],loc='upper left',facecolor='pink')
```

```python
a4=[10,20,30,40]
b4=[20,25,35,55]
fig=plt.figure(figsize=(6,4))
ax=fig.add_axes([1,1,1,1])
ax1=ax.plot(a4,b4,color='red')
ax2=ax.plot(b4,a4,color='#F2921D')
ax.set_title("Linear Graph",color='#301E67',fontsize=18)
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
ax.set_xticks(np.arange(5,65,step=5))
# ax.legend(labels=("Line 1","Line 2"),facecolor='#EFA3C8')
leg = plt.legend(labels=("Line 1","Line 2"),facecolor='lightblue')
text=leg.get_texts()
text[0].set_color("red")
text[1].set_color("#FF8B13")
```
